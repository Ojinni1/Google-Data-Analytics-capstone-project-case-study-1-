# Divvy Bikes

Divvy is Chicagoland’s bike share system across Chicago and Evanston. Divvy provides residents and visitors,
with a convenient, fun and affordable transportation option for getting around and exploring Chicago.

Divvy, like other bike share systems, consists of a fleet of specially-designed, sturdy and durable bikes 
that are locked into a network of docking stations throughout the region. The bikes can be unlocked from one 
station and returned to any other station in the system. People use bike share to explore Chicago, commute to 
work or school, run errands, get to appointments or social engagements, and more.

Divvy is available for use 24 hours/day, 7 days/week, 365 days/year, and riders have access to all bikes and 
stations across the system.
To read more check https://ride.divvybikes.com/about

Merge all data into one dataframe using concat, by using concat i'll be able to merge all data frames together without loosing any data and i'll be able to adjust the index 




```python
bike = pd.concat(objs= [data1,data2,data3,data4,data5,data6,data7,data8,data9,data10,data11,data12],
                 ignore_index=True)
```

```python
bike.isnull().sum() # checking for Nan Values in the dataframe
```

```python
# dropping Nan rows

bike = bike.dropna(how= 'any', axis = 0)

bike = bike.drop_duplicates(ignore_index= True)
```

```python
bike
```

The started_at and the ended_at  columns are in object, which needs to be changed to datetime to be able to read by python librbary this can also be done at the time of importing the dataset by using parse_date, but since we're merging all the dataframes into one lateron that's the reason it's being done here 

```python
bike['started_at'] = pd.to_datetime(bike['started_at'])

bike['ended_at'] = pd.to_datetime(bike['ended_at'])
```

```python
bike.head()
```

A column was created for the days of the week, where riders took the bike  

```python
bike['DayOfWeek'] = bike['started_at'].dt.day_name()
```
